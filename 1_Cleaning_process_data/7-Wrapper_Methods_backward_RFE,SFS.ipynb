```python
# read data
import numpy as np; import pandas as pd
import os
from google.colab import drive
drive.mount('/content/drive/')

%cd "/content/drive/My drive/hanhwa_AA/"

!git clone https://github.com/SungKu/Hanhwa_AA.git

linkpath = "/content/drive/My Drive/hanhwa_AA/"
dirname = os.getcwd()

data_pth = os.path.join(linkpath,'VSdata.csv')

VSdata = pd.read_csv(data_pth).values
```

```python
# separate X and y
y = VSdata[:,0]
X = VSdata[:,1:]
```

```python
# scale data
from sklearn.preprocessing import StandardScaler
xscaler = StandardScaler()
X_scaled = xscaler.fit_transform(X)

yscaler = StandardScaler()
y_scaled = yscaler.fit_transform(y[:,None])
```

```python
# Feature selection methods
from sklearn.feature_selection import SequentialFeatureSelector, RFE
from sklearn.linear_model import LinearRegression

# RFE-based feature selection
BRFE = RFE(LinearRegression(), n_features_to_select=10, step=1).fit(X_scaled, y_scaled)

# SFS-based feature selection
BSFS = SequentialFeatureSelector(LinearRegression(), n_features_to_select=10, direction='backward', cv=5).fit(X_scaled, y_scaled)

```

```python
# check selected inputs
print('Inputs selected by RFE: ', BRFE.get_support(indices=True)+1) # returns integer index of the features selected
print('Inputs selected by SFS: ', BSFS.get_support(indices=True)+1) # returns integer index of the features selected
```

```python
# reduce X to only top relevant inputs
X_relevant_RFE = BRFE.transform(X)
X_relevant_SFS = BSFS.transform(X)
```
