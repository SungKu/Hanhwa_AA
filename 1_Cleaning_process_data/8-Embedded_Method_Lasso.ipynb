```python
# read data
import numpy as np; import pandas as pd
import os
from google.colab import drive
drive.mount('/content/drive/')

%cd "/content/drive/My drive/hanhwa_AA/"

!git clone https://github.com/SungKu/Hanhwa_AA.git

linkpath = "/content/drive/My Drive/hanhwa_AA/"
dirname = os.getcwd()

data_pth = os.path.join(linkpath,'VSdata.csv')

VSdata = pd.read_csv(data_pth).values
```

```python
# separate X and y
y = VSdata[:,0]
X = VSdata[:,1:]
```

```python
# scale data
from sklearn.preprocessing import StandardScaler
xscaler = StandardScaler()
X_scaled = xscaler.fit_transform(X)

yscaler = StandardScaler()
y_scaled = yscaler.fit_transform(y[:,None])
```

```python
# fit Lasso model 
from sklearn.linear_model import LassoCV
Lasso_model = LassoCV(cv=5).fit(X_scaled, y_scaled.ravel())
```

```python
# cfind the relevant inputs using model coefficients
top_k_inputs = np.argsort(abs(Lasso_model.coef_))[::-1][:10] + 1
print('Relevant inputs: ', top_k_inputs)
```
