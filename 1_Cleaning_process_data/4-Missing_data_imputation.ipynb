```python
# Mean imputation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
import os
from google.colab import drive
drive.mount('/content/drive/')

%cd "/content/drive/My drive/hanhwa_AA/"

!git clone https://github.com/SungKu/Hanhwa_AA.git

linkpath = "/content/drive/My Drive/hanhwa_AA/"
dirname = os.getcwd()

data_pth = os.path.join(linkpath,'Load7.csv')

X = pd.read_csv(data_pth).values
X = X[:,1:3]
X[20000:23000,0] = np.nan

plt.plot(X[:,0], '-', alpha=0.3)
plt.xlabel('sample#'), plt.ylabel('value')
plt.title('Missing data'), plt.show()

mean_imputeModel = SimpleImputer(missing_values=np.nan, strategy='mean')

X_imputed = mean_imputeModel.fit_transform(X)

plt.plot(X[:,0], '-', alpha=0.3, label='Raw data')
plt.plot(np.arange(20000,23000),X_imputed[20000:23000,0], 'r-', alpha=0.7,label='Mean-imputed data')
plt.xlabel('sample#'), plt.ylabel('value'),plt.legend(loc='upper left')
plt.title('Missing data'), plt.show()
```

```python
# KNN imputation
from sklearn.impute import KNNImputer

knn_imputeModel = KNNImputer(n_neighbors=15)
X_impute_knn=knn_imputeModel.fit_transform(X)

plt.plot(X[:,0], '-', alpha=0.3, label='Raw data')
plt.plot(np.arange(20000,23000),X_imputed[20000:23000,0], 'r-', alpha=0.9,label='Mean-imputed data')
plt.plot(np.arange(20000,23000),X_impute_knn[20000:23000,0], 'm-', alpha=0.7,label='KNN-imputed data')
plt.xlabel('sample#'), plt.ylabel('value'),plt.legend(loc='upper left')
plt.title('Missing data'), plt.show()
```
